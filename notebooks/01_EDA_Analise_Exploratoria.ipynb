#### Separação das colunas por tipo

```python
coluna_target = ['Attrition']

colunas_categoricas_nao_ordenadas = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']

colunas_categoricas_ordinais = ['Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'PerformanceRating', 
                                'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance']

colunas_numericas = [coluna for coluna in df.columns if coluna not in (coluna_target + colunas_categoricas_nao_ordenadas + colunas_categoricas_ordinais)]
```

```python
colunas_numericas
```

#### Análises Gráficas - Colunas Numéricas

```python
fig, axs = plt.subplots(2, 7, figsize=(20, 10))

for ax, coluna in zip(axs.flatten(), colunas_numericas):
    sns.histplot(data=df, x=coluna, ax=ax, kde=True)

plt.tight_layout()

plt.show()
```

- Podemos ver que algumas features acima são de números inteiros apenas
- Também vemos muitas coluans com dados assimétricos (grande distribuição de dados a esquerda)

```python
fig, axs = plt.subplots(2, 7, figsize=(20, 10))

for ax, coluna in zip(axs.flatten(), colunas_numericas):
    sns.boxplot(data=df, x=coluna, ax=ax, showmeans=True)

plt.tight_layout()

plt.show()
```

```python
fig, axs = plt.subplots(2, 7, figsize=(20, 10))

for ax, coluna in zip(axs.flatten(), colunas_numericas):
    sns.boxplot(data=df, x=coluna, ax=ax, showmeans=True, hue='Attrition')

plt.tight_layout()

plt.show()
```

#### Análises Gráficas - Colunas Categóricas

```python
df[colunas_categoricas_nao_ordenadas].shape
```

```python
fig, axs = plt.subplots(2,4, figsize=(15,8), tight_layout=True)

for ax, coluna in zip(axs.flatten(), df[colunas_categoricas_nao_ordenadas]):
    df[coluna].value_counts().plot(kind='bar', ax=ax)

plt.show()
```

#### Análise de Correlação

```python
corr = df.select_dtypes('number').corr()
matriz = np.triu(df.select_dtypes('number').corr())

fig, ax = plt.subplots(figsize=(15,9), tight_layout=True)

sns.heatmap(corr, mask=matriz, annot=True, fmt='.2f', cmap='coolwarm', annot_kws={'fontsize': 9})

plt.show()
```

**Análise de Correlações**
- Não temos nenhuma correlação negativa forte
- As colunas relacionadas a "Anos" tem correlações maiores (Age, YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion ...)
- Também temos uma correlação entre JobLevel e MonthlyIncome que parece fazer sentido

#### Redução de tamanho da base de dados

```python
colunas_valores_inteiros = [coluna for coluna in df.select_dtypes('int')]

colunas_valores_inteiros
```

```python
df[colunas_valores_inteiros] = df[colunas_valores_inteiros].apply(pd.to_numeric, downcast='integer')

df.info()         # Veja que a base de dados saiu de 356 kB para 129 kB
```
